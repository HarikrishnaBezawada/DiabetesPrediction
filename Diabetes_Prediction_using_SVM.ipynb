```python
import pandas as pd
```

```python
import numpy as np
```

```python
file = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQRp2K84U9vl_vDKZMWfrh8Sx7R628KOJvTKiOMyaS3BP60rrr0XtnxyvWkm7QsNg/pub?output=csv")
```

```python
file.describe()
```

```python
file.isna().count()
```

```python
file.isnull().sum()
```

```python
features = file.rename(
    columns = {
        "Dataset for People for their Blood Glucose Level with their Superficial Body Feature readings. Age" : "Age",
        "Blood Glucose Level(BGL)" : "Glucose",
        "Diastolic Blood Pressure" : "Diastolic_BP",
        "Systolic Blood Pressure" : "Systolic_BP",
        "Sweating  (Y/N)" : "Sweating",
        "Shivering (Y/N)" : "Shivering",
        "Diabetic/NonDiabetic (D/N)" : "DiaOrNot"
    }
)
```

```python
features
```

```python
features.describe()
```

```python
FeaturesA = features.drop(columns={'DiaOrNot'})
```

```python
FeaturesA
```

```python
X = FeaturesA
```

```python
X
```

```python
Y = features['DiaOrNot']
```

```python
Y
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scalar = scalar.fit(X)
scalar
```

```python
Trans = scalar.transform(X)
Trans
```

```python
X = Trans
```

```python
X
```

```python
Y
```

```python
from sklearn import svm
```

```python
svm = svm.SVC(kernel = 'linear')
```

```python
svm.fit(X_train,Y_train)
```

```python
Predict = svm.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Output = accuracy_score(Predict,Y_train)
```

```python
Output
```

```python
svm.fit(X_test,Y_test)
```

```python
Test_Predict = svm.predict(X_test)
```

```python
Test_Predict
```

```python
Test_Out = accuracy_score(Test_Predict,Y_test)
```

```python
Test_Out
```

```python
Output = "{:.5f}".format(Output*100)
Test_Out = "{:.5f}".format(Test_Out*100)
```

```python
print("Accuracy score of training set is {a}% & testing data is {b}%" .format(b=Test_Out, a=Output))
```
